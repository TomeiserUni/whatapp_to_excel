```python
from pathlib import Path

BASE_DIR = Path("/Users/tomassilveira/Desktop/projeto/whatapp_to_excel")

CACHE_DIR = BASE_DIR / "cache"
PASTA_IMAGENS = BASE_DIR / "imagens_exemplo"
```

```python
import pickle
import numpy as np

def load_produtos():
    with open(CACHE_DIR / "prod.pkl", "rb") as f:
        produtos = pickle.load(f)
    embeddings = np.load(CACHE_DIR / "emb_prod.npy")
    return produtos, embeddings


def load_imagens():
    with open(CACHE_DIR / "img_text.pkl", "rb") as f:
        textos = pickle.load(f)
    embeddings = np.load(CACHE_DIR / "emb_img.npy")
    return textos, embeddings
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np

model = SentenceTransformer("all-MiniLM-L6-v2")


def cosine_similarity(a, b):
    return float(np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b)))
```

```python
def comparar_imagens_com_produtos():
    produtos, emb_prod = load_produtos()
    textos_img, emb_img = load_imagens()

    nomes_imagens = list(textos_img.keys())

    resultados = {}

    for nome_img, emb in zip(nomes_imagens, emb_img):
        scores = []

        for produto, emb_p in zip(produtos, emb_prod):
            score = cosine_similarity(emb, emb_p)
            scores.append((produto, round(score, 4)))

        scores.sort(key=lambda x: x[1], reverse=True)

        resultados[nome_img] = scores[:5]

    return resultados
```

```python
resultados = comparar_imagens_com_produtos()

for img, matches in resultados.items():
    print(f"\n {img}")
    for produto, score in matches:
        print(f" - {produto} ({score})")
```
