```python
from pathlib import Path
import easyocr

def images_to_dict(folder_path):
    reader = easyocr.Reader(['pt'])
    
    resultados = {}
    folder = Path(folder_path)

    for file in folder.iterdir(): 
        if file.suffix.lower() in [".png", ".jpg", ".jpeg"]:
            
            print(f" A processar: {file.name}")
            
            result = reader.readtext(str(file))  
            
            texto = " ".join([text for _, text, _ in result])
            
            resultados[file.name] = texto

    return resultados


```

```python
pasta = "/Users/tomassilveira/Desktop/projeto/whatapp_to_excel/imagens_treino"
dados = images_to_dict(pasta)

for img, texto in dados.items():
    print(f"\n{img} → {texto}")

```
